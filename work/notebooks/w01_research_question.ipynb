## 2. The question: decision, action, cost of a wrong call

### The Decision to Improve
Which articles should we update, expand, or optimize first this week?

### Who Acts on the Output & What They Do
SEO managers and editors will use this ranked list to assign updates to writers. They will check the flagged articles to update old stats, add new sections, or improve the keywords.

### Cost of a Wrong Call
- **Wrongly flagging a page (False Positive):** If we tell an editor to update a page that is actually doing fine and won't improve, we waste 2 to 4 hours of their time.
- **Missing a declining page (False Negative):** If we miss a page that is losing traffic, we lose clicks and views that could have been saved with a simple update.

### Why Data/ML Helps
Simple rules like "update if it is older than 6 months and gets 500 views" don't work well because page performance depends on a mix of things: where the page ranks on Google, the type of article, how old it is, and whether the click-through rate is dropping. A machine learning model can find the sweet spot between all these factors much better than a human writing manual rules.

## 3. Quick look at the data (2-3 real numbers)

I load the starter dataset `data/raw/content_refresh_anonymized.csv` to calculate key metrics that validate my lane choice. 

I look for:
1. **The overall decline rate:** What percentage of all pages in the dataset are actually losing traffic?
2. **Decline rates by Content Type:** Does the type of article affect how often it loses traffic?
3. **Decline rates by Position Tier:** Are pages in certain Google ranking positions more likely to decline?

```python
import pandas as pd
import numpy as np

# Load starter data
df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')
df['is_declining'] = df['trend_direction'].str.lower().eq('down').astype(int)

# 1. Overall decline rate
overall_decline_rate = df['is_declining'].mean()
print(f"1. Overall Decline Rate: {overall_decline_rate * 100:.1f}% ({df['is_declining'].sum()} out of {len(df)} pages)")

# 2. Decline rate by content type
decline_by_type = df.groupby('content_type')['is_declining'].mean() * 100
print("\n2. Decline Rate by Content Type:")
for ctype, pct in decline_by_type.items():
    print(f"   - {ctype}: {pct:.1f}%")

# 3. Decline rate by position tier
decline_by_tier = df.groupby('position_tier')['is_declining'].mean() * 100
print("\n3. Decline Rate by Position Tier:")
for tier, pct in decline_by_tier.items():
    print(f"   - {tier}: {pct:.1f}%")
```
